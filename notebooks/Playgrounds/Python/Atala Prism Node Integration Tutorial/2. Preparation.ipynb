# Preparation

This section explains how to prepare the environment to use the **Atala PRISM SDK**

## Prerequisites


#### System requirements:

- Any compatible operating system to use required tools & instruments
- At least 40 MB of free disk space for Atala PRISM SDK libraries

#### Tools & Instruments

- [Java 11](https://www.oracle.com/java/technologies/javase/jdk11-archive-downloads.html)
- [Python >= 3.7](https://www.python.org/downloads/)

---
**TIP**

For managing multiple Java versions we recommend using [SDKMAN!](https://sdkman.io/).

---

## Adding dependencies
You can add SDK modules to your project by downloading them from the GitHub Packages. See the next section for more details.

To be able to download packages:

1. Get Atala PRISM SDK GitHub authentication token from Atala team
2. Export `ATALA_GITHUB_TOKEN` environment variable with authentication token value

#### Use the following snippet to export a variable

Linux/MacOs
```bash
export ATALA_GITHUB_TOKEN="ghp_..."

```
Windows
```bash
set ATALA_GITHUB_TOKEN="ghp_..."

```

## Getting required libraries

To use Python wrapper for **Atala PRISM SDK**, it is required to download all dependencies first.

One way to do so is to use Atala PRISM CLI dependencies as it contains all required JAR files.

To download them, please, use the following command:

```bash
curl "https://maven.pkg.github.com/input-output-hk/atala-prism-sdk/io/iohk/atala/prism-cli/v1.3.3-snapshot-1657071792-bd25a29/prism-cli-v1.3.3-snapshot-1657071792-bd25a29.zip" -H "Authorization: Bearer ${ATALA_GITHUB_TOKEN}" -L -O
unzip prism-cli-v1.3.3-snapshot-1657071792-bd25a29.zip
```

Unzipped directory structure contains required `lib` directory with all JAR files.

---
**TIP**

If you already set up Kotlin project using **Atala PRISM SDK**, you could also add the following task to the `build.gradle.kts` script to save all dependencies in the build directory:.

```kotlin
val copyDependenciesToBuildDir by tasks.creating(Copy::class) {
    from(configurations.runtimeClasspath.get().resolve())
    into(layout.buildDirectory.dir("dependencies"))
    dependsOn(tasks.build)
}
```

Then, save dependencies to `<project_dir>/build/dependencies` using `./gradlew copyDependenciesToBuildDir` command.

---

Then, set `JAVA_HOME` and `ATALA_PRISM_JARS` environment variables as follows:

```bash
export JAVA_HOME=<java_home_directory>
export ATALA_PRISM_JARS="<working_dir>/prism-cli-v1.3.3-snapshot-1657071792-bd25a29/lib"
```

Install [JPype](https://jpype.readthedocs.io/en/latest/) to access Java classes from Python:

```bash
pip install -U pip setuptools wheel
pip install JPype1==1.3.0
```

---
**INFO**

**Atala PRISM SDK** has no Python implementation.

Python is used as a wrapper for Atala PRISM SDK. Under the hood, it uses JPype to get access to Java classes of SDK from Python.

For Java classes imported with JPype indexation is not supported.
