## A
### agent
An agent or digital agent is software. This software allows an entity to perform actions, communicate, and store information.

### Anti Money Laundering
Anti Money Laundering (AML) is a set of laws and regulations to prevent the disguising of illegal funds. This can include montoring of suspicious transactions or financial activity. Activities can include, money laundering, terrorist financing, and other financial crimes.

## C
### claim
An authoritative statement of fact about a [subject](#subject).


### credential
A [verifiable credential](#verifiedcredential) (VC) is tamper-resistant document that can be cryptographically verified. VC's can be used for [verifiable presentations](#verifiablepresentation), which are also crytopgraphically verifiable. The [claim](#claim) within a credential can be about various [subjects](#subject).


## D
### decentralized identifier
A decentralized identifier (DID), is a globally unique persistent identifier that requires no centralized registration authority and can be generated and/or registered cryptographically. The basic format of a DID is defined by the [DID syntax](#didsyntax). The [DID scheme](#didscheme) is defined in the [DID method](#didmethod) specification. PRISM DIDs make use of distributed ledger technology (DLT)](#distributedledgertechnology).


### decentralized identity management
Identity management that uses [decentralized identifiers](#decentralizedidentifier). It extends authority for DID generation, registration, and assignment outside of traditional trusted directory services. 


### DID
Acronym for [decentralized identifier](#decentralizedidentifier)


### DID controller
An entity that can make changes to the [DID document](#diddocument). It is possible to have more than one DID controller, the [DID subject](#didsubject) may be the controller. The controller can be found in the ```controller``` property at the top of the [DID document](#diddocument).


### DID document
Data that describes the [DID subject](#didsubject), including public keys that the [DID subject](#didsubject) can authenticate itself and prove association with the [DID](#did). 


### DID fragment
A portion of a [DID URL](#didurl) that follows the first hash sign character(#). The DID fragment syntax copies the URI fragment syntax.


### DID method
Outlines how a [DID method scheme](#didscheme) is implemented. A DID method specification defines which operations create, resolvee, update and revoke [DIDs](#did) and [DID documents](#diddocument).


### DID path
The snippet of a [DID URL](#didurl) beginning with and including the forward slash (/) and ends with either a question mark (?), fragment hash (#), or the end of the [DID URL](#didurl). The syntax for the DID path is identical to the URI path.


### DID query
The snippet of a [DID URL](#didurl) that begins with and includes the first question mark (?). The syntax for the DID query is identical to the URI query.


### DID resolution
A process that takes a [DID](#did) as input and resolution options. It returns a [DID document](#diddocument) in a standard template, plus any additional metadata. This process is dependent upon on the 'Read' operation of the [DID method](#didmethod).


### DID resolver
Is software and/or a hardware component that performs the [DID resolution](#didresolution).


### DID scheme
Formalized syntax for a [decentralized identifier](#decentralizedidentifier). The basic scheme begins with prefix ```did``` defined in the [DID syntax](#didsyntax). Each [DID method](#didmethod) has a scheme that works with that specific implementation. In a scheme, the DID method name follows the URI: ```did:prism```.


### DID subject
The entity that is identified by a [DID](#did) and described by a [DID document](#diddocument). Any entity can be a subject: person, group, organization, corporation, physical or digital items, etc.


### DID syntax
The basic DID scheme is a URI scheme that conforms ot the [DID Syntax ABNF Rules](https://datatracker.ietf.org/doc/html/rfc5234).


### DID URL
A [DID](#did) plus any required syntax components, including an optional [DID path](#didpath) with the leading forward slash (/), optional [DID query](#didquery) with the leading question mark (?), and optional [DID fragment](#didfragment) with leading hash character (#).


### distributed ledger technology
A distributed ledger technology (DLT) is a decentralized system ledger for recording events. DLT's use distributed databases and consensus protocols to confirm the ordering of cryptographically signed transactions. Linking the digitally signed transactions over time makes the history of the ledger immutable.


## E
### entity
A thing with a distinct existstence, i.e. a person, corporation, organization, physical or digital thing, or device that have one or more roles in the ecosystem.

## G
### Governance framework
A set of business, legal and technical rules make up a governance framework. An example, credit card networks like Visa and Mastercard have established rules for usage of cards.


## H
### holder
[Entities](#entity) may hold one or more [verifiable credentials](#verifiedcredential) and generate a [presentation](#presentation) from the VCs they are holding. Holders store credentials in a repository (wallet).


## I
### issuer
An [entity](#entity) can assert [claims](#claim) about one or more [subjects](#subject), creating [verifiable credentials](#verifiedcredential) from these claims and *issuing* the VC to a [holder](#holder).

## K
### Know Your Customer (KYC)
A regulation and/or ethical requirement for banking and securities industries that requires institutions to verify the identity of customers and assess risk factors. This may include elements of [AML](#antimoneylaundering).

## M
### Merkle tree
Merkle trees are cryptographic data structures. They provide a compact and efficient way to verify the integrity of large data sets. Blockchain and decentralized technologies have adopted Merkle trees as a core component. The basic concept is that they can provide mathematical proof that a specific item, like a blockchain transaction, exists somewhere in the blockchain (dataset).

## P
### presentation
Data extracted from a [verified credential](#verifiedcredential) that is shared with a [verifier](#verifier). 


### public key description
A data element inside of a [DID document](#diddocument) with the required metadata to use a public or verification key.


## Q
### QR code
The QR codes referenced would contain a connection invitation. This invitation would request information from the individual who scanned the code. This request would establish a connection or prove an existing connection. These QR codes wouldn't be secret or need strong security in the encrypted connections between users or issuer and holder.

## R
### repository
A program like a storage vault or wallet that stores and secures [verifiable credentials](#verifiedcredential).


### resource
The term resource is anything identified by a URI. This term is interchangable with [entity](#entity).


## S
### selective disclosure
The ability to make decisions about what data that a [holder](#holder) shares with another entity. For example, sharing my first name, but not my full name.


### self-sovereign identity
Self-sovereign identity (SSI) is a new model for identity in Web3. It allows us to prove our identity, establish trusted (secure) relationships with other entities to share or protect information. SSI is driven by new techologies in cryptography, blockchains, cloud computing and smart phones.

### SSI
Acronym for [self-sovereign identity](#selfsovereignidentity).


## T
### Trust framework
Trust frameworks are also known as [governance frameworks](#governanceframework).


### trust registry
A list of trusted members that form a network. This network can be a *source of truth* for knowing who can issue specific credentials, i.e., an education board issues a *credential* of accrediation to a school/education organization.

## U
### Uniform Resource Identifier (URI)
Standard identifer on the internet. A [DID] is one of many URI schemes alongside http, ftp, and smb. 


### Uniform Resource Locator (URL)
A way to specify a location of a [resource](#resource) or [entity](#entity). This is the same thing as a web address, i.e., https://iohk.io. An URL can also apply to a [DID](#did).


## V
### verified credential
Verified credential (VC) is a data model for cryptographically-verifiable credentials. 


### verification
The evaluation of the authenticity of a [verifiable credential](#verifiedcredential) or [verifiable presentation](#verifiablepresentation) from an [issuer](#issuer) or presenter ([holder](#holder)). 


### verifier
An [entity](#entity) performs a [verification](#verification) by being presented one or more VCs. 


### verifiable presentation
Tamper-resistant presentation that is encrypted in such a way that authorship of the data is trustworthy through crytographic verification. This can be done in a synthesized form that do not contain the original VC, i.e., [zero-knowledge-proofs](#zeroknowledgeproofs).


### zero-knowledge-proof
A zero-knowledge proof (ZKP) is a cryptographic method where an entity can prove to another entity that they know a specific value without disclosing the actual value.
